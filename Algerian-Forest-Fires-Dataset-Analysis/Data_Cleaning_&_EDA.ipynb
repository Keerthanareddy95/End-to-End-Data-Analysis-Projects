```python
from ast import increment_lineno
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
dataset.head()
```

```python
dataset.info()
```

**Data Cleaning**

```python
# missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
df = dataset
df.loc[:123,'Region']=0
df.loc[123:,'Region']=1
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
df.head()
```

```python
df.isnull().sum()
```

```python
# removing null
df = df.dropna().reset_index(drop= True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# removing 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fixing spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
# changing the required columns as int
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
# changing other columns to float data type other than classes
objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
for i in objects:
  if i!= 'Classes':
    df[i]=df[i].astype(float)
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# saving the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned.csv',index = False)
```
