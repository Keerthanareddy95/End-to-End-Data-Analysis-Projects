```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter plot of Score vs. Popularity
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Score', y='Popularity', data= anime_df, alpha=0.5)
plt.title('Score vs. Popularity')
plt.xlabel('Score / Rating of the Anime')
plt.ylabel('Popularity')
plt.show()
```

The animes with lower score (rating) , tend to have lesser popularity as many data points are scattered towards larger ranking number.

```python
# Histogram of Scores
plt.figure(figsize=( 6,4))
plt.hist(anime_df['Score'], bins=30, color='yellow', edgecolor='black')
plt.title('Distribution of Anime Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

This distribution shows most of the animes released have an average rating between 7.8 and 8.2

```python
anime_df['Type'].nunique()
```

```python
# Count plot of Anime Types
plt.figure(figsize=( 6,4))
sns.countplot(x='Type', data= anime_df, palette='viridis')
plt.title('Distribution of Anime Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

This plot shows that most of the animes released belong to the category of "TV" and least number of them belong to the "Special" category!

```python
# Extract and count genres
from collections import Counter
genres_count = Counter([genre for sublist in anime_df['Genres'].str.split(', ') for genre in sublist])

# Plot top genres
plt.figure(figsize=(6, 4))
top_genres = dict(genres_count.most_common(10))  # Get top 10 genres
sns.barplot(x=list(top_genres.values()), y=list(top_genres.keys()), palette='pastel')
plt.title('Top 10 Anime Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

This plot clearly describes that the most commom genre is Action and the top 10 genres include Adventure, Fantasy, Drama, Romance, Supernatural, Mystery, Sci-Fi and Suspense..!!

```python
# Pie chart of Ratings
plt.figure(figsize=(6, 4))
anime_df['Rating'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('Distribution of Anime Ratings')
plt.ylabel('')
plt.show()
```

This pie chart Visulaizes and gives a summary that PG-13 - Teens or older category of animes have the highest ratings and thus, are most popular among the audience!

```python
# Extract and count producers
producers_count = Counter([producer for sublist in anime_df['Producers'].str.split(', ') for producer in sublist])

# Plot top producers
plt.figure(figsize=(6,4))
top_producers = dict(producers_count.most_common(10))  # Get top 10 producers
sns.barplot(x=list(top_producers.values()), y=list(top_producers.keys()), palette='rocket')
plt.title('Top 10 Anime Producers')
plt.xlabel('Count')
plt.ylabel('Producer')
plt.show()
```

This graph clearly describes that the top producer is Shueisha!

```python
# Create a cross-tabulation of Status and Type
status_type_cross = pd.crosstab(anime_df['Type'], anime_df['Status'])

# Plot stacked bar chart
plt.figure(figsize=(6,4))
status_type_cross.plot(kind='bar', stacked=True, colormap='Set3')
plt.title('Distribution of Status by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Status')
plt.show()
```

This stacked bar graph describes the status of the animes based, which are classified based on their type.

```python
# Extract year from 'Premiered' column
anime_df['Premiered_Year'] = anime_df['Premiered'].str[-4:].astype(int)

# Count number of premieres per year
premiered_counts = anime_df['Premiered_Year'].value_counts().sort_index()

# Plot number of premieres over years
plt.figure(figsize=(6,4))
sns.lineplot(x=premiered_counts.index, y=premiered_counts.values, marker='o', color='skyblue')
plt.title('Number of Anime Premiered Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Premieres')
plt.xticks(rotation=45)
plt.show()
```

This lineplot shows the number of animes released over the years.
We can see a sharp rise in the number after 2020 which signifies the current popularity of anime across all the age groups and genres.

```python
# Scatter plot of Members vs. Popularity
plt.figure(figsize=(6,4))
sns.scatterplot(x='Members', y='Popularity', data= anime_df, alpha=0.5)
plt.title('Members vs. Popularity')
plt.xlabel('Members')
plt.ylabel('Popularity')
plt.show()
```

This plot shows that the animes having more number of associated members have the most popularity and we can also observe a clear pattern among the relationship between the number of associated memebers and popularity of the anime.

```python
# Sort the DataFrame by 'Rank' to get the top 10 anime titles
top_10_anime = anime_df.sort_values(by='Rank').head(10)

# Extract English titles and corresponding ranks
titles = top_10_anime['English']
ranks = top_10_anime['Rank']

# Plotting the top 10 anime titles based on rank
plt.figure(figsize=(6,4))
plt.barh(titles, ranks, color='skyblue')
plt.xlabel('Rank')
plt.ylabel('Anime Title (English)')
plt.title('Top 10 Anime Titles Based on Rank')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest rank at the top
plt.show()
```

Here are the top 10 animes displayed based on their ranking!
